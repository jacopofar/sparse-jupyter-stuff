This notebook import TSV data inside a local Postgres database. It assumes the credentials are the one in the first lines:

```python
import psycopg2
conn = psycopg2.connect(dbname="postgres", user="postgres", password="mysecretpassword", host="localhost")
# should be UTF-8
print('connection encoding, used as default:', conn.encoding)
```

```python
cur = conn.cursor()
cur.execute("CREATE TABLE IF NOT EXISTS chat_messages (message TEXT, message_time TIMESTAMP, username TEXT, channel TEXT);")
cur.execute("TRUNCATE TABLE chat_messages;")
conn.commit()
cur.close()
```

```python
import time
cur = conn.cursor()
counter = 0
start_time = time.time()
for l in open('/Users/utente/IdeaProjects/IngestAdiumLogs/adium_chat_messages.tsv'):
    counter = counter + 1
    if counter % 5000 == 0:
        print(f'inserted {counter} lines')
        conn.commit()
    parts = l.split('\t')
    channel = parts[0]
    ts = parts[1]
    username = parts[2]
    
    username = username[:50]
    channel = channel[:50]
    
    msg_body = parts[3]
    cur.execute("INSERT INTO chat_messages (message, message_time, username, channel) VALUES (%s, %s, %s, %s)", (msg_body, ts, username, channel))

conn.commit()

elapsed_time = round(time.time() - start_time)
print(f'process completed,  inserted {counter} lines in {elapsed_time}')
```

```python
start_time = time.time()

cur.execute("CREATE INDEX channel_idx1 ON chat_messages(channel);")
conn.commit()

cur.execute("CREATE INDEX channel_time_idx1 ON chat_messages(channel ASC, message_time ASC);")
conn.commit()

elapsed_time = time.time() - start_time
print(f'indexes created on {elapsed_time} seconds')
```
